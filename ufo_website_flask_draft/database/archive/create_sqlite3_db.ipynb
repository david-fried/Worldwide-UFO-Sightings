```python
db_conn = sqlite3.connect('../database.sqlite')
```

```python
c = db_conn.cursor()
```

```python
c.execute('DROP TABLE IF EXISTS ufo_data;')

c.execute(
    
"""
CREATE TABLE ufo_data(
id SERIAL PRIMARY KEY,
datetime VARCHAR,
city VARCHAR,
state VARCHAR,
country VARCHAR,
shape VARCHAR,
duration FLOAT,
duration_hours VARCHAR,
comments VARCHAR,
date_posted VARCHAR,
latitude FLOAT,
longitude FLOAT,
year FLOAT,
month FLOAT,
Foreign Key (state) REFERENCES state_stats(abbreviation));

"""
    
)
```

```python
c.execute('DROP TABLE IF EXISTS state_stats;')

c.execute(
    
"""

CREATE TABLE state_stats(
name VARCHAR,
drug_deaths INT,
abbreviation VARCHAR,
sightings INT,
avg_sighting_duration FLOAT,
code VARCHAR,
census_area FLOAT,
PRIMARY KEY (abbreviation));

"""
    
)
```

```python
c.execute('DROP TABLE IF EXISTS merge_again;')

c.execute(
    
"""
CREATE TABLE merge_again(
year FLOAT,
death_rate FLOAT,
drug_deaths VARCHAR,
state_codes FLOAT,
state_names VARCHAR,
state_census_area FLOAT,
state_abb VARCHAR,
state_sightings FLOAT,
duration FLOAT,
changing FLOAT,
chevron FLOAT,
cigar FLOAT,
circle FLOAT,
cone FLOAT,
cross FLOAT,
cylinder FLOAT,
diamond FLOAT,
disk FLOAT,
egg FLOAT,
fireball FLOAT,
flash FLOAT,
formation FLOAT,
light FLOAT,
other FLOAT,
oval FLOAT,
rectangle FLOAT,
sphere FLOAT,
teardrop FLOAT,
triangle FLOAT,
unknown FLOAT,
id INT PRIMARY KEY,
state VARCHAR,
abbr VARCHAR,
poverty FLOAT,
povertyMoe FLOAT,
age FLOAT,
ageMoe FLOAT,
income FLOAT,
incomeMoe FLOAT,
healthcare FLOAT,
healthcareLow FLOAT,
healthcareHigh FLOAT,
obesity FLOAT,
obesityLow FLOAT,
obesityHigh FLOAT,
smokes FLOAT,
smokesLow FLOAT,
smokesHigh FLOAT,
GEO_ID VARCHAR,
NAME VARCHAR,
population INT);

"""
)
```

```python
c.execute('DROP TABLE IF EXISTS shape_counts;')

c.execute(
    
"""

CREATE TABLE shape_counts(
State VARCHAR PRIMARY KEY,
Changing FLOAT,
Chevron FLOAT,
Cigar FLOAT,
Circle FLOAT,
Cone FLOAT,
Cross FLOAT,
Cylinder FLOAT,
Diamond FLOAT,
Disk FLOAT,
Egg FLOAT,
Fireball FLOAT,
Flash FLOAT,
Formation FLOAT,
Light FLOAT,
Other FLOAT,
Oval FLOAT,
Rectangle FLOAT,
Sphere FLOAT,
Teardrop FLOAT,
Triangle FLOAT,
Unknown FLOAT);

"""
    
)
```

```python
ud.to_sql('ufo_data', db_conn, if_exists='append', index=False)
ss.to_sql('state_stats', db_conn, if_exists='append', index=False)
ma2.to_sql('merge_again', db_conn, if_exists='append', index=False)
bd.to_sql('shape_counts', db_conn, if_exists='append', index=False)
```

```python
# Test with SQLAlchemy

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from scipy.stats import ttest_ind
```

```python
engine = create_engine("sqlite:///../database.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
ufo_data = Base.classes.ufo_data
state_stats = Base.classes.state_stats
merge_again = Base.classes.merge_again
shape_counts = Base.classes.merge_again
```

```python
session = Session(engine)
```

```python
sightings = session.query(state_stats.sightings)
```

```python
drugs_data = session.query(state_stats.name, state_stats.drug_deaths).all()
```
