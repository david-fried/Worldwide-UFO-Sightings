```python
#dependencies
import pandas as pd
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
from nltk.probability import FreqDist

ufo_data_load = "data/scrubbed.csv"
```

```python
#read csv
ufo_data = pd.read_csv(ufo_data_load, low_memory=False)
```

```python
ufo_data_comments = ufo_data['comments']
```

```python
ufo_data_comments
```

```python
#drop null values
ufo_data['comments'].dropna(inplace=True)
tokens = ufo_data['comments'].apply(word_tokenize)
```

```python
tokens
```

```python
tokens=list(ufo_data['comments'].unique())
indicators= [tokens[0]]
for i in range(1, len(tokens)-1):
    if tokens[i+1].split()[0:2] != tokens[i].split()[0:2]:
        indicators.append(tokens[i+1])
```

```python
str1 = ''.join(indicators)
```

```python
words = word_tokenize(str1)
```

```python
fdist = FreqDist(words)
```

```python
fdist.most_common(200)
```

```python
import matplotlib.pyplot as plt

fdist.plot(30)
```

```python
#nltk.download('stopwords')
```

```python
words_no_punc = []

for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())
```

```python
from nltk.corpus import stopwords

stopwords = stopwords.words("english")
```

```python
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)
```

```python
#jupyter notebook --NotebookApp.iopub_data_rate_limit=1.0e10
#jupyter lab --NotebookApp.iopub_data_rate_limit=1.0e10
```

```python
listToStr = ' '.join([str(elem) for elem in clean_words])
```

```python
fdist = FreqDist(clean_words)
```

```python
top_50 = fdist.most_common(50)
```

```python
fdist.plot(15)
```
