```python
import pandas as pd
```

```python
df1 = pd.read_csv('Source/source_drug_deaths.csv')
df2 = pd.read_csv('Source/source_sightings.csv')
ufo_data = pd.read_csv('ufo_data.csv')
```

```python
ufo_data
```

```python
ufo_data.nunique()
```

```python
ufo_data.drop_duplicates(subset='datetime', keep='first', inplace=True)
```

```python
df1.Indicator.value_counts()
```

```python
ods = df1[df1.Indicator == 'Number of Drug Overdose Deaths']
ods
```

```python
state_drug_deaths = ods[(ods.Year==2019) & (ods.Month=='January')].reset_index(drop=True)
state_drug_deaths['Data Value'] = state_drug_deaths['Data Value'].str.replace(',', '')
state_drug_deaths = state_drug_deaths.astype({'Data Value': 'int64'})
state_drug_deaths.dtypes
```

```python
state_drug_deaths = state_drug_deaths[['State Name','Data Value']].groupby('State Name')['Data Value'].sum()
state_drug_deaths
```

```python
state_stats = pd.merge(state_drug_deaths, df2, on="State Name")
state_stats.rename(columns={'Data Value': 'Drug Deaths', 'Duration': 'AvgSightingDuration', 'State': 'State Abbreviation'}, inplace=True)
state_stats.to_csv('state_stats.csv', index=None)
state_stats
```

```python
ufo_data.rename(columns={'state': 'State Abbreviation'}, inplace=True)

state_data_full = pd.merge(state_stats, ufo_data, on='State Abbreviation')

state_data_full.to_csv('state_data_full.csv', index=None)

state_data_full.rename(columns={'Sightings': 'State Sightings'}, inplace=True)

state_data_full
```
