```python
import pandas as pd
import numpy as np
```

```python
ufo_data = pd.read_csv('ufo_data.csv', low_memory=False)
```

```python
def year(x):
    if '/' in x:
        a = x.find('/')
        b = x[a+1:].find('/')
        return x[a+b+2: a+b+6]
    else:
        x = np.nan

ufo_data['year'] = ufo_data.datetime.apply(year)

ufo_data.year.dropna(inplace=True)

ufo_data.year = ufo_data.year.astype('int64')

ufo_data.year.dropna(inplace=True)

ufo_data.year
```

```python
def month(x):
    if '/' in x:
        a = x.find('/')
        return x[:a]
    else:
        x = np.nan

ufo_data['month'] = ufo_data.datetime.apply(month)

ufo_data.month.dropna(inplace=True)

ufo_data.month = ufo_data.month.astype('int64')

ufo_data.month.dropna(inplace=True)

ufo_data.month
```
